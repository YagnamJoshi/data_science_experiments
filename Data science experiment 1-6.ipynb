### removing outliers

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```

## Experiment 3
### Title: Data Preprocessing-Applying, encoding techniques using data set obtained from  UCI ML repository

### Import encoding libraries

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

### Add encoders

```python
label_encoder = LabelEncoder()
#onehot_encoder = OneHotEncoder(sparse=False)
```

### Perform label encoding on class

```python
df['CLASS'] = label_encoder.fit_transform(df['CLASS'])
df.head()
```

### Perform one hot encoding 

```python
onehot_encoder_shape = OneHotEncoder(sparse=False)
shape_encoded = onehot_encoder_shape.fit_transform(df[['SHAPE']])
shape_encoded_df = pd.DataFrame(shape_encoded, columns=onehot_encoder_shape.get_feature_names_out(['SHAPE']))

# Initialize OneHotEncoder for 'SOIL_TYPE'
onehot_encoder_soil = OneHotEncoder(sparse=False)
soil_type_encoded = onehot_encoder_soil.fit_transform(df[['SOIL_TYPE']])
soil_type_encoded_df = pd.DataFrame(soil_type_encoded, columns=onehot_encoder_soil.get_feature_names_out(['SOIL_TYPE']))

# Concatenate the original dataframe with the one-hot encoded columns
df_encoded = pd.concat([df, shape_encoded_df, soil_type_encoded_df], axis=1)
df_encoded.drop(['SHAPE', 'SOIL_TYPE'], axis=1, inplace=True)
df_encoded.head()
```

```python
### Spliting the dataset
```

```python
from sklearn.model_selection import train_test_split
X = df_encoded.drop(columns=['CLASS'])  # Features
y = df_encoded['CLASS']  # Target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Checking null values in splited dataset
print("Y test:",y_train.isnull().sum())
print("Y train:",y_test.isnull().sum())
```

```python
# Treating them
# Calculate the mean of the non-null values in y_train
mean_y_train = y_train.mean()

# Replace NaN values in y_train with the calculated mean
y_train.fillna(mean_y_train, inplace=True)

# Calculate the mean of the non-null values in y_test
mean_y_test = y_test.mean()

# Replace NaN values in y_test with the calculated mean
y_test.fillna(mean_y_test, inplace=True)
```

```python
# Checking null values again
print("Y test:",y_train.isnull().sum())
print("Y train:",y_test.isnull().sum())
```
